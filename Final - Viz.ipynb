```python

# Specify the path to your JSON file
file_path = 'results/bigger_p4.json'

# Open the JSON file and load its contents
with open(file_path, 'r') as file:
    bigger_plot_data = json.load(file)


```

```python
variances=[]
for b in bigger_plot_data:
    d={}
    for runs,vals in b.items():
        d[runs]={}
        for metric,values in vals["debug"].items():
#            print(metric,values)
            d[runs][metric]={}

            d[runs][metric]=np.var(values)
    variances.append(d)
```

```python
vars_df=pd.DataFrame()
for idx,i in enumerate(variances):
    print(vars_df)
    i_df=pd.DataFrame(i)
    i_df.reset_index(inplace=True)
    i_df["nrs"]=idx
   # print(i_df)
    vars_df=pd.concat([vars_df, i_df], ignore_index=True)
```

```python
hhhj=pd.melt(vars_df,id_vars=["nrs","index"],value_vars=["run 0","run 1","run 2","run 3","run 4","run 5","run 6"])
```

```python
hhhj.replace( "run ","",regex=True,inplace=True)
hhhj["nrs"]=hhhj["nrs"]*5+5
```

```python
hhhj.columns=["Nr of repetitions","Metric","Nr of columns mutated","Variance"]
```

```python
sns.color_palette()
```

```python
sns.set(rc={'figure.figsize':(7,5)})
hhhj.replace({"cross":"CC","rbo":"RBO","ndgc_score":"NDCG","r2_score":"R2","cohen_kappa_score":"Cohen K","kendalltau":"Ktau","levenshtein_normalized_similarity":"Lev","jaro_winkler_normalized_similarity":"Jaro","weightedtau":"WKtau","damerau_levenshtein_normalized_similarity":"Damerau","hamming_normalized_similarity":"Hamming"},inplace=True)

g = sns.FacetGrid(hhhj, col="Metric",col_wrap=3,sharey=True,sharex=False,legend_out=True)
g.map(sns.scatterplot, "Nr of columns mutated", "Variance","Nr of repetitions")
g.add_legend()
g.tight_layout()
g.savefig("Viz/facet_plot.png",dpi=600)
```
