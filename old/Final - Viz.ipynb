```python
# Define a function that will be applied
def create_cc_rows(row, name):
    # Find a match in df2 based on the 'value' column of df1
    vals = []
    dt = row["CC"][name]
    if name == "aggregated":
        vals = row["CC"][name]
    else:
        for k, v in dt.items():
            if v[0] == np.inf:
                vals.append(0)
            else:
                vals.append(v[0])

    return np.mean(vals)
```

```python
import pickle

with open("results/synth_result.pkl", "rb") as f:
    synth_result = pickle.load(f)
data_synth = {k: [] for k in synth_result[0]["aggregated"].keys()}
for i in synth_result:
    for k, v in i["aggregated"].items():
        data_synth[k].append(v)
bardata = pd.DataFrame(data_synth)
bardata["CC"] = [i["cross"] for i in synth_result]

# Create a new column in df1 using apply and lookup_function
bardata["real_synth"] = bardata.apply(create_cc_rows, axis=1, name="real_synth")
bardata["synth_real"] = bardata.apply(create_cc_rows, axis=1, name="synth_real")
bardata["CC"] = bardata.apply(create_cc_rows, axis=1, name="aggregated")

bardata
```

```python
import numpy as np
```

```python
variances = []
for _, b in bigger_plot_data_full.items():
    d = {}

    for runs, vals in b.items():
        d[runs] = {}
        for metric, values in vals.items():
            # print(metric,values)
            d[runs][metric] = {}

            d[runs][metric] = np.var(values)
    variances.append(d)
```

```python
import pandas as pd
```

```python
vars_df = pd.DataFrame()
for idx, i in enumerate(variances):
    # print(vars_df)
    i_df = pd.DataFrame(i)
    i_df.reset_index(inplace=True)
    i_df["nrs"] = idx
    # print(i_df)
    vars_df = pd.concat([vars_df, i_df], ignore_index=True)
```

```python
hhhj = pd.melt(
    vars_df,
    id_vars=["nrs", "index"],
    value_vars=["run 0", "run 1", "run 2", "run 3", "run 4", "run 5", "run 6"],
)
hhhj.replace(
    {
        "cross": "CC",
        "rbo": "RBO",
        "ndgc_score": "NDCG",
        "r2_score": "R2",
        "cohen_kappa_score": "Cohen K",
        "kendalltau": "Ktau",
        "levenshtein_normalized_similarity": "Lev",
        "jaro_winkler_normalized_similarity": "Jaro",
        "weightedtau": "WKtau",
        "damerau_levenshtein_normalized_similarity": "Damerau",
        "hamming_normalized_similarity": "Hamming",
    },
    inplace=True,
)
```

```python
hhhj.replace("run ", "", regex=True, inplace=True)
hhhj["nrs"] = hhhj["nrs"] * 5 + 5
```

```python
hhhj.columns = ["Nr of repetitions", "Metric", "Nr of columns mutated", "Variance"]
```

```python
crosstab1 = pd.crosstab(
    index=[hhhj["Nr of repetitions"], hhhj["Nr of columns mutated"]],
    columns=[hhhj["Metric"]],
    values=hhhj["Variance"],
    aggfunc="mean",
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame (Assume you already have df_combined)

# Add a column to group the row indices into groups of 5 (First group by rows)
df_combined["Group"] = df_combined.index // 5  # Integer division to group every 5 rows

# Define custom group names based on index groups (first group by 5 rows)
group_names = {
    0: "Group 1 (Rows 0-4)",
    1: "Group 2 (Rows 5-9)",
    2: "Group 3 (Rows 10-14)",
    3: "Group 4 (Rows 15-19)",
    4: "Group 5 (Rows 20-24)",
}
df_combined["Group"] = df_combined["Group"].map(group_names)

# Add a second group for marking based on (0,5,10), (1,6,11), etc.
df_combined["Marker Group"] = (
    df_combined.index % 5
)  # This creates the pattern (0,5,10), (1,6,11), etc.

# Define custom names for the marker group
marker_group_names = {0: "heart", 1: "breast", 2: "thyroid", 3: "liver", 4: "tumour"}
df_combined["Marker Group"] = df_combined["Marker Group"].map(marker_group_names)

# Convert the DataFrame from wide-form to long-form for plotting
df_long = pd.melt(
    df_combined,
    id_vars=["Group", "Marker Group"],
    var_name="Metric",
    value_name="Value",
)

# Define custom colors for the groups
custom_palette = {
    "Group 1 (Rows 0-4)": "#FF5733",
    "Group 2 (Rows 5-9)": "#33FF57",
    "Group 3 (Rows 10-14)": "#3357FF",
    "Group 4 (Rows 15-19)": "#F0A500",
    "Group 5 (Rows 20-24)": "#9B59B6",
}

# Set the figure size before plotting
plt.figure(figsize=(12, 8))  # Adjust the size (width, height) as needed

# Create the violin plot (without the marker columns)
sns.violinplot(x="Metric", y="Value", data=df_long, inner=None, split=True)

# Overlay with a scatterplot using custom colors (hue) and markers (style)
sns.scatterplot(
    x="Metric",
    y="Value",
    data=df_long,
    hue="Group",
    style="Marker Group",
    palette=custom_palette,
    s=100,
    alpha=0.7,
)

# Set labels and rotate x-axis ticks
plt.xlabel("Metric")
plt.ylabel("Value")
plt.xticks(rotation=90)

# Set y-axis limits (optional)
plt.ylim(0, 1)  # Adjust as needed

# Adjust the legend to be more informative
plt.legend(title="Row Group & Marker Group", bbox_to_anchor=(1.05, 1), loc="upper left")

# Save the figure to a file
plt.savefig("Viz/synthetic_violin_scatter_custom_markers.png", bbox_inches="tight")

# Show the plot (optional)
plt.show()
```
